```python
# Rename the columns so they follow the PE8 (snake case)
def to_snake_case(text):
    result = [text[0].lower()]
    for char in text[1:]:
        if char.isupper():
            result.extend(['_', char.lower()])
        else:
            result.append(char)
    return ''.join(result)

# Get the existing column names and convert them to snake_case
employee.columns = [to_snake_case(column) for column in employee.columns]
employee
```

```python
# Create and test (at least one) hypothesis
```

```python
# Hypothesis: "The attrition rate in the company is below 15%"
# H0: The attrition rate is 15% or higher
# H1: The attrition rate is below 15%
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
expected_proportion = 0.15
attrition_count = len(employee_data[employee_data['attrition'] == 'Yes'])
total_employee = len(employee_data)
stat, p_value = proportions_ztest(attrition_count, total_employee, expected_proportion)
print(stat, p_value)
```

use the 0.95 for the confidence level

p_value > 0.05 and stat is greater than 0, therefore we fail to reject null hypothesis null.

There is not enough evidence to suggest that the attrition rate is below 15%

```python
# Hypothesis: Gender has significant effect on employee attrition
# H0: There is no significant difference in attrition rates between male and female employees
# H1: Gender significantly affects employee attrition, and there is a difference in attrition rates between male and female employees.
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table for gender and attrition
contingency_table = pd.crosstab(employee_data['gender'], employee_data['attrition'])

# Perform a chi-squared test to determine if gender and attrition are independent
chi2, p, _, _ = chi2_contingency(contingency_table)
print(p)

# Check the p-value to determine statistical significance
if p < 0.05:  # use 0.95 for the confidence level
    print("Hypothesis: Gender significantly affects employee attrition (Reject Null Hypothesis)")
else:
    print("Hypothesis: Gender has no significant effect on employee attrition (Fail to Reject Null Hypothesis)")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, G
radientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def test_models(X_train, X_test, y_train, y_test, models=None):
    models = {
        'Logistic Regression': LogisticRegression(),
        'KNN': KNeighborsClassifier(),
        'MLP Classifier': MLPClassifier(max_iter=10000),
        'Random Forest': RandomForestClassifier(),
        'Gradient Boosting': GradientBoostingClassifier(),
        'SVC': SVC(),
        'Decision Tree': DecisionTreeClassifier(),
        'Naive Bayes': GaussianNB()
    }

    results = {}

    for name, model in models.items():
        model.fit(X_train, y_train)  # train the model
        predictions = model.predict(X_test)  # predict on the test set

        results[name] = {
            "Accuracy": accuracy_score(y_test, predictions),
            "Precision": precision_score(y_test, predictions),
            "Recall": recall_score(y_test, predictions),
            "F1 Score": f1_score(y_test, predictions)
        }
       
        print(f"Results for {name}:")
        print(results[name])
        print('-' * 40)

    return results
```

```python
results = test_models(X_train_SMOTE,X_test_transformed,y_train_SMOTE,y_test)
```
